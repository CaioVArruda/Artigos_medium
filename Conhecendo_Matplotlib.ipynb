# Importando as bibliotecas

Aqui vamos importar as duas bibliotecas que iremos utilizar nesse tutorial incial, o Matplotlib assunto principal e o Numpy que utilizaremos para gerar os dados demonstrativos para essa nossa discussão.

```python
#Importando as bibliotecas que vamos utilizar
import matplotlib.pyplot as plt
import numpy as np
```

# Plotando os gráficos

A partir deste tópico vamos elaborar vários gráficos exemplicando quais comandos podemos executar para visualizá-los e até mesmo como estilizar alguns parãmetros.

```python
# Estamos criando uma figura onde o gráfico será plotado, o fig será responsável direto por isso
fig = plt.figure()

# Estamos determinando os nossos eixos x e y que formaram uma base para que os dados sejam plotados na figura construida acima
ax = plt.axes()

# Vamos usar o comando abaixo, pois este nos fornecerá números igualmente espaçados em um determinado intervalo
x = np.linspace(0, 10, 1000)# intervalo entre 0 e 10

# Comando específico para plotar o gráfico
ax.plot(x, np.sin(x));# Veja que x representa a nossa base de dados criada no eixo x e no eixo y utilizamos outra ferramenta do numpy que representa a função seno baseada em x
```

Uma outra forma de plotar o gráfico e por meio do comando plt, o qual faz referência a biblioteca matplotlib quando a importamos para o nosso ambiente de trabalho.

```python
plt.plot(x, np.sin(x));
```

# Plotando seno e cosseno

Repare que podemos plotar vários gráficos em uma mesma figura, veja como é legal a sobreposição do seno e do cosseno onde conseguimos observar a defasagem de fase que existe entre essas duas funções  e o comportamento sinoidal que ambas possuem. Outro detalhe é a cor, neste caso as cores foram impostas conforme o padrão da própria biblioteca.

```python
# Gráfico do seno e cosseno
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

# Alterando as cores

É possível configurar a cor da linha que une todos os pontos que foram plotados no nosso gráfico, através do parâmetro color ou c e atribuir a ele a cor que você deseja. A lista de cores é possível descobrir na documentação da biblioteca, esta pode ser inserida de diversas maneiras as quais serão representadas abaixo.

```python
# Plotando linhas de diferentes cores e chamando cores das mais diversas maneiras
plt.plot(x, x-0, color='blue')
plt.plot(x, x-1, color='g')
plt.plot(x, x-2, color='0.75')
plt.plot(x, x-3, color='#FFDD44')
plt.plot(x, x-4, color=(1.0, 0.2, 0.3))
plt.plot(x, x-5, color='chartreuse');
```

```python
# Plotando seno com deslocamento de uma unidade entre eles
plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='g')
plt.plot(x, np.sin(x-2), color='0.75')
plt.plot(x, np.sin(x-3), color='#FFDD44')
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3))
plt.plot(x, np.sin(x-5), color='chartreuse');
```

# Alterando o estilo da linha

Assim como as cores é possível alterar o estilo da linha por do comando linestyle e atribuir um dos argumentos suportados por este parâmetro, perceba que no exemplo abaixo utilizamos alguns dos mais comuns, entre eles o sólido, tracejado, tracejado com ponto e o pontilhado.

```python
# Alterando o estilo da linha
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted');
```

Temos um outro jeito de atribuir o estilo da linha ao invés de passar o nome destes podemos simplesmente inserir alguns caracteres especiais como solid = '_'; dashed = '--'; dashdot = '-.' e dotted = ':'.

```python
# Outra forma de inserir argumentos para alterar o estilo da linha
plt.plot(x, x+0, linestyle='-')
plt.plot(x, x+1, linestyle='--')
plt.plot(x, x+2, linestyle='-.')
plt.plot(x, x+3, linestyle=':');
```

# Alterando cor e estilo no mesmo argumento

Podemos alterar os dois parâmetros por meio de um único argumento, na primeira linha de comando temos o argumento '-g' onde o tracejado representa o estilo da linha, ou seja, sólido e a letra g representa a cor que no caso é o green (verde).

```python
# Alterando cor e estilo no mesmo argumento
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--c')
plt.plot(x, x+2, '-.k')
plt.plot(x, x+3, ':r');
```

# Modificando limites

Também é possível alterar tanto o limite superior e inferior, aumento os eixos do gráfico no sentido de x e de y. Esse objetivo é atingido por meio do comando plt.yLim ou .xlim, conforme exemplo abaixo.

```python
# Plotando a função seno
plt.plot(x, np.sin(x))

# Estabelecendo limites superior e inferior para o gráfico
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

O comando **axis** permite que modifiquemos os limites do gráfico de forma direta inserido os limites como argumento e em forma de lista como podemos observar abaixo, exemplo ([xmin, xmax, ymin, ymax]), além deste podemos inserir outros como option que possue como argumento a seguinte relação:

* **on:** Torna os eixos e linhas vísiveis pode ser substituido por True;
* **off:** Torna os eixos e linhas invisiveis pode ser substituido por False; 
* **equal:** Configura a escala de maneira simétrica, podendo incorrer em mudanças nos limites dos eixos, estes podem não ser respeitados;
* **scaled:** Pode alterar as dimensões da figura;
* **tight:** Configura os limites suficientes para mostrar todos os dados;

Temos mais alguns argumentos que podem ser vistos na documentação do [matplotlib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html).





```python
# Plotando a função seno
plt.plot(x, np.sin(x))

# Estabelecendo limites em x e y por meio do axis
plt.axis([-1, 11, -1.5, 1.5])

# determinando o enquadramento por meio do argumento option
plt.axis('tight');
```

# Adicionando título e legenda

Para podermos arrumar o layout do nosso gráfico podemos adicionar título por meio do comando title e legendas por meio do comando legend. Dentro deste comando ainda temos outros argumentos que podem ser alterados com a finalidade de dar uma aparência mais clean para o gráfico, como posição, tamanho, cor, estilo da fonte entre outras características.

```python
# Adicionando o título
plt.title('sin and cos')

# Veja que dentro do nosso plot adicionamos o label o qual integrará a legenda
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), '--r', label='cos(x)')

# determinando o enquadramento por meio do argumento option
plt.axis('equal')

# Atribuindo a legenda no nosso gráfico
plt.legend();
```

Finalizando temos uma utilizando todos os comandos que aprendemos acima.

```python
# Adicionando o título
plt.title('sin e cos')

# Plotando a função seno
plt.plot(x, np.sin(x-0), '-b', label='sin(x-0)')
plt.plot(x, np.sin(x-1), '--g', label='sin(x-1)')
plt.plot(x, np.sin(x-2), '-.r', label='sin(x-2')
plt.plot(x, np.sin(x-3), ':c', label='sin(x-3)')
plt.plot(x, np.sin(x-4), '-k', label='sin(x-4)')
plt.plot(x, np.sin(x-5), '--y', label='sin(x-5)')

# determinando o enquadramento por meio do argumento option
plt.axis('tight')

# Atribuindo a legenda no nosso gráfico
plt.legend();
```
