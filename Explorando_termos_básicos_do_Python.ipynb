
*   Compara√ß√£o: tem grande utilidade para elaborarmos certas circunst√¢ncias (condi√ß√µes) e como o pr√≥prio nome diz comparar duas vari√°veis entre si. Para os exemplos vamos a = 10 e b = 20.

Cada objeto criado dentro do Python pode ser identificado de duas maneiras seja por um n√∫mero de identifica√ß√£o por meio do comando .id(), sendo esse o endere√ßo do objeto na mem√≥ria, ou pode ser identificado atrav√©s da sua classifica√ß√£o dentre alguns tipos existentes em python, como lista, tupla, dicion√°rio, string, inteiro, float, por meio do comando type().

**Dicion√°rio**:

Estes dados possuem keys e values que est√£o conectados entre si, no sistema tipo chave fechadura, como exemplo podemos mencionar um diret√≥rio telef√¥nico que possue diversos campos, dentre esses queremos linkar o campo nome como key a um value como dado, ou seja, temos um par dados ligados entre si.

Para elaborarmos um dicion√°rio precisamos partir do princ√≠pio de que precisamos atribuir esse dicion√°rio a uma vari√°vel da seguintes maneira, e = dict() ou e = {}. Dentro destas chaves teremos : a esquerda fica a nossa key e a sua direita o nosso value. A key pode ser uma string ou um inteiro.

```python
# Exemplo de um dicion√°rio

telefone = {'nome':['Caio', 'Melina', 'Isadora'], 'telefone':['12345', '123456', '12345678'], 'email':['@gmail.com', '@hotmail.com', '@yahoo.com']}

# Este comando imprimi na tela tanto as keys como os values, ou seja, o dicion√°rio inteiro.
print(telefone)

# Se quisermos acessar as keys de maneira individualizada usaremos o seguinte comando
print(telefone['nome'])
print(telefone['telefone'])
print(telefone['email'])

# Al√©m destes comandos citados acima temos alguns outros.

print(telefone.keys())
print(telefone.values())

print(telefone.get('nome'))

print(telefone.items())

tel = telefone.copy()

print(tel)

telefone.update(a='Gustavo')

print(telefone)
```

**Condicionais de declara√ß√£o**:

Servem para executar uma declara√ß√£o/statement ou grupo de declara√ß√£o/statement, podemos nomear 3 condi√ß√µes (if, elif, else);

2022-08-25.png

*   O algoritmo come√ßa verificando a primeira condi√ß√£o, se True executa a statement a;
*   Se a primeira condi√ß√£o for False o algoritmo se encaminha para a segunda condi√ß√£o (elif), se True executa a statement b;
*   Entretanto, se a segunda condi√ß√£o for False o programa executar√° a terceira condi√ß√£o (else);





```python
# Exemplo de como funciona o processo dos condicionais e como o algoritmo vai verificando cada uma delas.

x = 10
y = 10

if x < y:
  print('x √© menor que y.')
elif x > y:
  print('x √© maior que y.')
else:
  print('x √© ig y.')
```

**Loops**:

Acabamos de ver os condicionais e neles percebemos que h√° uma sequ√™ncia l√≥gica de execu√ß√£o, entretanto, existe uma situa√ß√£o na qual precisamos um bloco de comando v√°rias vezes, para que isto ocorra utilizamos o loop, este como papel executar esse bloco in√∫meras vezes at√© que n√£o seja mais necess√°rio.

Loop.png

*   Primeiro ocorre a verifica√ß√£o da condi√ß√£o, se True o loop √© executado, colocando em execu√ß√£o o bloco que est√° sob sua responsabilidade;
*   Executou uma primeira vez esse bloco, retorna a condi√ß√£o e verifica novamente, se True executa o loop mais uma vez;
*   Essas etapas acima ocorreram enquanto a condi√ß√£o continuar retornando True, no momento em que retornar False o loop n√£o √© mais executado e o algoritmo parte para o restante do c√≥digo;

o loop pode ser classificado em infinito, quando a condi√ß√£o nunca retornar√° False e finito, a condi√ß√£o em uma dada circunst√¢ncia retornar√° False e loop n√£o ser√° executado.

Temos tr√™s tipos:

**While**:

Usado quando n√£o se sabe ao certo quantas vezes precisar√° executar o looping. 


```python
# Exemplo de Looping while

count = 0

while(count < 10):
  print(count)
  count = 1 + count # ou count += 1

print('Fui.')
```

**For**:

Permite que o bloco de comandos seja repetido um certo n√∫mero de vezes, ent√£o sabemos exatamente quantas vezes o algoritmo passar√° por esse bloco comandos espec√≠fico.

```python
# Exemplo de condicional for

nome = ['Caio', 'Melina', 'Isadora']

for i in nome:
  print(i, end=',')
```

**Nested**:

Podemos dizer que √© um looping dentro do outro, ou seja, temos um while dentro do for e vice-versa, ou while dentro do while ou for dentro do for.

```python
# Exemplo nested

count = 1

for i in range(10):
  print(str(i)*i)
  
  for j in range(0, i):
    count = count + 1
```

Por fim vamos comentar o que s√£o fun√ß√µes, que de um jeito pr√°tico podem ser definidas como uma pr√°tica usada para dividir o c√≥digo em blocos √∫teis, tornando-o mais leg√≠vel.

```python
# Exemplo de fun√ß√£o

def soma (a, b):
  return a + b

c = soma(10, 20)
print(c)
```

No exemplo acima definimos uma fun√ß√£o de soma, assim toda vez que quisermos somar valores basat chamar esta fun√ß√£o e lhe fornecer os argumentos que deseja somar e pronto. Assim, percebemos como este pode ser √∫til para deixar o nosso c√≥digo mais leg√≠vel conforme a sua confec√ß√£o.
